```python
# Import Dependencies
import pandas as pd
import numpy as np
```

## Import All Colony Data

```python
csv_path = "Data/Colony_Data/hcny_2015_1.csv"
colony_2015_1_df = pd.read_csv(csv_path)
colony_2015_1_df.head()
```

```python
# colony_2015_1_df.dtypes
```

```python
csv_path = "Data/Colony_Data/hcny_2015_2.csv"
colony_2015_2_df = pd.read_csv(csv_path)
colony_2015_2_df.head()
```

```python
csv_path = "Data/Colony_Data/hcny_2015_3.csv"
colony_2015_3_df = pd.read_csv(csv_path)
colony_2015_3_df.head()
```

```python
csv_path = "Data/Colony_Data/hcny_2015_4.csv"
colony_2015_4_df = pd.read_csv(csv_path)
colony_2015_4_df.head()
```

## Merge ALL Colony DataFrames and Calculate Sum

```python
sum_2015_df = pd.concat((colony_2015_1_df, colony_2015_2_df, colony_2015_3_df, colony_2015_4_df))
sum_2015_df
```

```python
colony_2015_df = sum_2015_df.groupby("State").sum()
colony_2015_df.head()
```

```python
# Remove Percent_Lost and Percent_Renovated Columns
colony_2015_df = colony_2015_df[['Colony_Number', 'Max_Colony_Number', 'Lost_Colony_Number',
                                 'Added_Colony_Number', 'Renovated_Colony_Number']]
colony_2015_df
```

## Import All Bee Stressor Data

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2015_1.csv"
stressor_2015_1_df = pd.read_csv(csv_path)
stressor_2015_1_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2015_2.csv"
stressor_2015_2_df = pd.read_csv(csv_path)
stressor_2015_2_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2015_3.csv"
stressor_2015_3_df = pd.read_csv(csv_path)
stressor_2015_3_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2015_4.csv"
stressor_2015_4_df = pd.read_csv(csv_path)
stressor_2015_4_df.head()
```

## Merge ALL Stressor DataFrames and Calculate Mean

```python
stressor_all_df = pd.concat((stressor_2015_1_df, stressor_2015_2_df, stressor_2015_3_df, stressor_2015_4_df))
stressor_all_df
```

```python
stressor_2015_df = stressor_all_df.groupby("State").mean()
stressor_2015_df.head()
```

## Merge ALL DataFrames for 2015

```python
csv_path = "Data/Production_Data/honey_2015.csv"
honey_2015_df = pd.read_csv(csv_path)
honey_2015_df.head()
```

```python
merge_1_df = pd.merge(honey_2015_df, colony_2015_df, on=["State"], how="outer")
merge_1_df.head()
```

```python
data_2015_df = pd.merge(merge_1_df, stressor_2015_df, on=["State"], how="outer")
data_2015_df
```

```python
# csv files 
data_2015_df.to_csv('data_2015.csv', index=False, header=True)
```
