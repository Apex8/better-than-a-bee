```python
# Import Dependencies
import pandas as pd
import numpy as np
```

## Import All Colony Data

```python
csv_path = "Data/Colony_Data/hcny_2018_1.csv"
colony_2018_1_df = pd.read_csv(csv_path)
colony_2018_1_df.head()
```

```python
# colony_2018_1_df.dtypes
```

```python
csv_path = "Data/Colony_Data/hcny_2018_2.csv"
colony_2018_2_df = pd.read_csv(csv_path)
colony_2018_2_df.head()
```

```python
csv_path = "Data/Colony_Data/hcny_2018_3.csv"
colony_2018_3_df = pd.read_csv(csv_path)
colony_2018_3_df.head()
```

```python
csv_path = "Data/Colony_Data/hcny_2018_4.csv"
colony_2018_4_df = pd.read_csv(csv_path)
colony_2018_4_df.head()
```

## Merge ALL Colony DataFrames and Calculate Sum

```python
sum_2018_df = pd.concat((colony_2018_1_df, colony_2018_2_df, colony_2018_3_df, colony_2018_4_df))
sum_2018_df
```

```python
colony_2018_df = sum_2018_df.groupby("State").sum()
colony_2018_df.head()
```

```python
# Remove Percent_Lost and Percent_Renovated Columns
colony_2018_df = colony_2018_df[['Colony_Number', 'Max_Colony_Number', 'Lost_Colony_Number',
                                 'Added_Colony_Number', 'Renovated_Colony_Number']]
colony_2018_df
```

## Import All Bee Stressor Data

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2018_1.csv"
stressor_2018_1_df = pd.read_csv(csv_path)
stressor_2018_1_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2018_2.csv"
stressor_2018_2_df = pd.read_csv(csv_path)
stressor_2018_2_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2018_3.csv"
stressor_2018_3_df = pd.read_csv(csv_path)
stressor_2018_3_df.head()
```

```python
csv_path = "Data/Bee_Stressor_Data/Stressor_2018_4.csv"
stressor_2018_4_df = pd.read_csv(csv_path)
stressor_2018_4_df.head()
```

## Merge ALL Stressor DataFrames and Calculate Mean

```python
stressor_all_df = pd.concat((stressor_2018_1_df, stressor_2018_2_df, stressor_2018_3_df, stressor_2018_4_df))
stressor_all_df
```

```python
stressor_2018_df = stressor_all_df.groupby("State").mean()
stressor_2018_df.head()
```

## Merge ALL DataFrames for 2018

```python
csv_path = "Data/Production_Data/honey_2018.csv"
honey_2018_df = pd.read_csv(csv_path)
honey_2018_df.head()
```

```python
merge_1_df = pd.merge(honey_2018_df, colony_2018_df, on=["State"], how="outer")
merge_1_df.head()
```

```python
data_2018_df = pd.merge(merge_1_df, stressor_2018_df, on=["State"], how="outer")
data_2018_df
```

```python
# csv files 
data_2018_df.to_csv('data_2018.csv', index=False, header=True)
```
