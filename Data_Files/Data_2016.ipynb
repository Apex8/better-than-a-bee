## Merge ALL Colony DataFrames and Calculate Sum

```python
sum_2016_df = pd.concat((colony_2016_1_df, colony_2016_2_df, colony_2016_3_df, colony_2016_4_df))
sum_2016_df
```

```python
colony_2016_df = sum_2016_df.groupby("State").sum()
colony_2016_df.head()
```

```python
# Remove Percent_Lost and Percent_Renovated Columns
colony_2016_df = colony_2016_df[['Colony_Number', 'Max_Colony_Number', 'Lost_Colony_Number',
                                 'Added_Colony_Number', 'Renovated_Colony_Number']]
colony_2016_df
```

## Import All Bee Stressor Data

## Merge ALL Stressor DataFrames and Calculate Mean

```python
stressor_all_df = pd.concat((stressor_2016_1_df, stressor_2016_2_df, stressor_2016_3_df, stressor_2016_4_df))
stressor_all_df
```

```python
stressor_2016_df = stressor_all_df.groupby("State").mean()
stressor_2016_df.head()
```
