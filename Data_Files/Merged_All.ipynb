```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
csv_path = "Data/Merged_Year_Data/data_2015.csv"
data_2015_df = pd.read_csv(csv_path)
data_2015_df.head()
```

```python
csv_path = "Data/Merged_Year_Data/data_2016.csv"
data_2016_df = pd.read_csv(csv_path)
data_2016_df.head()
```

```python
csv_path = "Data/Merged_Year_Data/data_2017.csv"
data_2017_df = pd.read_csv(csv_path)
data_2017_df.head()
```

```python
csv_path = "Data/Merged_Year_Data/data_2018.csv"
data_2018_df = pd.read_csv(csv_path)
data_2018_df.head()
```

```python
csv_path = "Data/Merged_Year_Data/data_2019.csv"
data_2019_df = pd.read_csv(csv_path)
data_2019_df.head()
```

```python
merge_all_df = pd.concat((data_2015_df, data_2016_df, data_2017_df, data_2018_df, data_2019_df))
merge_all_df
```

```python
merge_all_df = merge_all_df.sort_values(['State'])
merge_all_df
```

```python
# csv files 
merge_all_df.to_csv('merge_all.csv', index=False, header=True)
```
